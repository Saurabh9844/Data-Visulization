# Q.3 Read motorcycle and classic car productline sales data and show it using a multiline plot and take first 100 rows from dataframe.Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).



```python
df["PRODUCTLINE"].value_counts()
```

```python
ne=df.head(100)
mt=ne[ne["PRODUCTLINE"]=="Motorcycles"]
cc=ne[ne["PRODUCTLINE"]=="Classic Cars"]
x1=mt["MONTH_ID"]
y1=mt["SALES"]
x2=cc["MONTH_ID"]
y2=cc["SALES"]
plt.figure(figsize=(15,8))
plt.plot(x1,y1,marker="o",color='b',linestyle='dashed',label="Motorcycles Sales Data")
plt.plot(x2,y2,marker="o",color='g',linestyle='dashed',label="Classic Cars sales Data")
plt.xlabel("MONTH NUMBER")
plt.ylabel("SALES")
plt.title("Company Sales Data")
plt.legend(loc="best")
```

## 3: Read motorcycle and classic car productline sales data and show it using a multiline plot and take first 100 rows from dataframe .
Display the number of units sold per month for each product using multiline plots. (i.e., Separate 
Plotline for each product for each product)

```python
new=data.head(100)
Motorcycle=new[new["Productline"]=="Motorcycles"]
Classic_car=new[new["Productline"]=="Classic Cars"]
x1=Motorcycle["MONTH_ID"]
y1=Motorcycle["SALES"]
x2=Classic_car["MONTH_ID"]
y2=Classic_car["SALES"]
plt.figure(figsize=(10,5))
plt.plot(x1,y1,marker="o",color="b",linestyle="-",label="Motorcycles Sales Data")
plt.plot(x2,y2,marker="^",color="g",linestyle="-",label="Classic Cars Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("company Sales Data")
plt.legend(loc="best")
```

## 5: Read Classic Cars and motorcycle data product sales data and show it using the bar chart and take first 100 rows from data frame.

## 8: Calculate total sale data for motor cycle and classic car product and show it using a Pie chart take first 100 rows from dataframe

```python
plt.figure(figsize=(14,7))
Y1=y1.sum()
Y2=y2.sum()
plt.pie([Y1,Y2],labels=["Motorcycle","classic cars"],autopct="%1.2f%%",shadow=True)
plt.legend(loc="lower right")
plt.show()
```

## 9: Read motorcycle and classic car of all months and display it using the Subplot and take first 100 rows from data frame

```python
x1=Motorcycle["MONTH_ID"]
y1=Motorcycle["SALES"]
x2=Classic_car["MONTH_ID"]
y2=Classic_car["SALES"]

plt.figure(figsize=(12,9))
plt.subplot(3,1,1) 
plt.plot(x1,y1,color="r",marker="*",label="Motorcycle sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Data of Motorcycle")
plt.legend()

plt.figure(figsize=(12,9))
plt.subplot(3,1,2)  
plt.plot(x2,y2,color="g",marker="*",label="Classic cars sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Data of Classic cars")
plt.legend()

plt.figure(figsize=(12,9))
plt.subplot(3,1,3)  
plt.bar(x1,y1,label="Motorcycle sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Barplot of Motorcycle")
plt.legend()
```

## 10: Calculate total sale data for motor cycle ,classic car, truck and bus, vintage car product and show it using a Pie chart take first 1000 rows from data-frame

```python
rows=data.head(1000)
new=rows.groupby("PRODUCTLINE")
Motorcycle=new.get_group("Motorcycles")
Classic_car=new.get_group("Classic Cars")
Vintage_car=new.get_group("Vintage Cars")
truks=new.get_group("Trucks and Buses")
x3=Motorcycle["SALES"].sum()
x4=Classic_car["SALES"].sum()
x5=Vintage_car["SALES"].sum()
x6=truks["SALES"].sum()

plt.figure(figsize=(14,7))
plt.pie([x3,x4,x5,x6],labels=["Motorcycle","Classic_cars","Vintage_cars","Trucks and Buses"],autopct="%1.2f%%")
plt.legend(loc="lower right")
plt.show()
```
